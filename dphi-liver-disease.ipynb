```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('../input/no-title/train_dataset.csv')
train.head()
```

```python
test = pd.read_csv('../input/no-title/test_dataset.csv')
test.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train = train.fillna(train.mode().iloc[0])
train.head()
```

```python
train.isnull().sum()
```

```python
test = test.fillna(test.mode().iloc[0])
test.head()
```

```python
test.shape
```

```python
test.isnull().sum()
```

```python
import category_encoders as ce
ce_OHE = ce.OneHotEncoder(cols=['Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema'])

train = ce_OHE.fit_transform(train)
train.head()
```

```python
test = ce_OHE.fit_transform(test)
test.head()
```

```python
X = train
y = test
```

```python
X_train = X.iloc[:, :-1]
y_train = X.iloc[:,-1]
```

```python
X_test = y
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
  

```

```python
import xgboost as xg
from sklearn.metrics import mean_squared_error as MSE
from sklearn import datasets, linear_model, metrics


reg = linear_model.LinearRegression()
 
# train the model using the training sets
reg.fit(X_train, y_train)

 
# regression coefficients
print('Coefficients: ', reg.coef_)
 
```

```python
import matplotlib.pyplot as plt
```

```python

# Instantiation
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 123)
  
# Fitting the model
xgb_r.fit(X_train, y_train)
  
# Predict the model
pred = xgb_r.predict(X_test)
  
```
