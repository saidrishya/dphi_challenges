```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('/content/Train_data.csv')
train_data.head()
```

```python
test_data = pd.read_csv('/content/Test_data.csv')
test_data.head()
```

```python
print(train_data.shape)
print('..................')
print(test_data.shape)
```

```python
print(train_data.isnull().sum().sum())
print('...................')
print(test_data.isnull().sum().sum())
```

```python
print(train_data.isnull().value_counts())
print('...................')
print(test_data.isnull().value_counts())
```

```python
x = train_data['Soil_Type_18'].isnull().value_counts()
len(x)
```

```python
for i in train_data:
  x = train_data[i].isnull().value_counts()
  if len(x) > 1:
    train_data[i].fillna(train_data[i].mode()[0], inplace=True)
```

```python
print(train_data.isnull().value_counts())
```

```python
for i in test_data:
  x = test_data[i].isnull().value_counts()
  if len(x) > 1:
    test_data[i].fillna(test_data[i].mode()[0], inplace=True)
```

```python
print(test_data.isnull().value_counts())
```

```python
train_data.describe()
```

```python
test_data.describe()
```

```python
X = train_data.drop('Cover_Type', axis = 1)
y = train_data['Cover_Type']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validate, y_train, y_validate = train_test_split(X,y,test_size = 0.25, random_state=42)
```

```python
X_test = test_data
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_validate)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_validate,pred )
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators = 100)
random_forest.fit(X_train, y_train)
```

```python
random_pred = random_forest.predict(X_validate)
```

```python
accuracy_score(y_validate,random_pred )
```

```python
from sklearn.naive_bayes import GaussianNB
naive_model = GaussianNB()
naive_model.fit(X_train, y_train)
```

```python
naive_pred = naive_model.predict(X_validate)
```

```python
accuracy_score(y_validate,naive_pred )
```

```python
y_pred = random_forest.predict(X_test)
```
